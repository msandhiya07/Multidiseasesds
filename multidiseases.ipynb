```python
text_columns=['red_blood_cell_count','packed_cell_volume','white_blood_cell_count']
for i in text_columns:
  print(f"{i} : {k_df[i].dtype}")
```

```python
def convert_text_to_numeric(k_df, column):
    k_df[column] = pd.to_numeric(k_df[column], errors="coerce")

for column in text_columns:
    convert_text_to_numeric(k_df, column)
    print(f"{column}: {k_df[column].dtypes}")
```

```python
missing=k_df.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
def mean_inputation(k_df,column):
  mean=k_df[column].mean()
  k_df[column].fillna(value=mean_value,inplace=True)

def mode_inputation(k_df,column):
  mode=k_df[column].mode()[0]
  k_df[column].k_df[column].fillna(mode)
```

```python
def mean_value_imputation(df, column):
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# Select numeric columns
num_cols = [col for col in k_df.columns if k_df[col].dtype != 'object']

# Apply mean imputation to each numeric column
for col_name in num_cols:
    mean_value_imputation(k_df, col_name)
    print(f"{col_name}: Imputation done")
```

```python
# Define mode imputation function
def mode_value_imputation(df, column):
    mode_value = df[column].mode()[0]  # Get the most frequent value (mode)
    df[column].fillna(mode_value, inplace=True)

# Identify categorical columns
cat_cols = [col for col in k_df.columns if k_df[col].dtype == 'object']

# Apply mode imputation to each categorical column
for col_name in cat_cols:
    mode_value_imputation(k_df, col_name)
    print(f"{col_name}: Mode imputation done")
```

```python
# Define mapping for categorical columns
categorical_mappings = {
    'classification': {'ckd': 1, 'notckd': 0},
    'hypertension': {'yes': 1, 'no': 0},
    'diabetes_mellitus': {'yes': 1, 'no': 0},
    'coronary_artery_disease': {'yes': 1, 'no': 0},
    'appetite': {'good': 1, 'poor': 0},
    'pedal_edema': {'yes': 1, 'no': 0},
    'anemia': {'yes': 1, 'no': 0}
}

# Apply mapping to each categorical column
for col, mapping in categorical_mappings.items():
    k_df[col] = k_df[col].map(mapping)
    print(f"{col}: {k_df[col].unique()}")

```

```python
from sklearn.model_selection import train_test_split  # Import train_test_split

# Split features and target
x = k_df.drop("classification", axis=1)
y = k_df["classification"]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=25)

# Print shapes of the datasets
print(f"'x' shape: {x.shape}")
print(f"'y' shape: {y.shape}")
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and fit the model
gnb = GaussianNB()
gnb.fit(x_train, y_train)

# Create a list of models
models = [gnb]

# Append another model with a tuple (name, instance)
models.append(("Naive Bayes", GaussianNB()))
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python

from sklearn.ensemble import RandomForestClassifier

models = [
    ('Naive Bayes', GaussianNB())
]

for name, model in models:
    model.fit(x_train, y_train)  # Fit model inside the loop
    y_pred = model.predict(x_test)

    print(name, model)
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python
p_df= pd.read_csv('/content/parkinsons - parkinsons.csv')
```

```python
print(p_df.head())
print(p_df.info())
```

```python
X = p_df.drop(['status', 'name'], axis=1)
y = p_df['status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
from xgboost import XGBClassifier

# Remove the 'use_label_encoder' parameter, as it's not needed in the latest versions of XGBoost
xgb_model = XGBClassifier(eval_metric='logloss')
xgb_model.fit(X_train, y_train)

# Make predictions and calculate accuracy for XGBoost
y_pred_xgb = xgb_model.predict(X_test)

# Accuracy for XGBoost
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")

# Classification Report and Confusion Matrix for XGBoost
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
print("XGBoost Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```
