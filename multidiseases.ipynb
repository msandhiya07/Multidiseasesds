```python
def convert_text_to_numeric(k_df, column):
    k_df[column] = pd.to_numeric(k_df[column], errors="coerce")

for column in text_columns:
    convert_text_to_numeric(k_df, column)
    print(f"{column}: {k_df[column].dtypes}")
```

```python
def mean_inputation(k_df,column):
  mean=k_df[column].mean()
  k_df[column].fillna(value=mean_value,inplace=True)

def mode_inputation(k_df,column):
  mode=k_df[column].mode()[0]
  k_df[column].k_df[column].fillna(mode)
```

```python
def mean_value_imputation(df, column):
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# Select numeric columns
num_cols = [col for col in k_df.columns if k_df[col].dtype != 'object']

# Apply mean imputation to each numeric column
for col_name in num_cols:
    mean_value_imputation(k_df, col_name)
    print(f"{col_name}: Imputation done")
```

```python
# Define mode imputation function
def mode_value_imputation(df, column):
    mode_value = df[column].mode()[0]  # Get the most frequent value (mode)
    df[column].fillna(mode_value, inplace=True)

# Identify categorical columns
cat_cols = [col for col in k_df.columns if k_df[col].dtype == 'object']

# Apply mode imputation to each categorical column
for col_name in cat_cols:
    mode_value_imputation(k_df, col_name)
    print(f"{col_name}: Mode imputation done")
```

```python
p_df= pd.read_csv('/content/parkinsons - parkinsons.csv')
```

```python
print(p_df.head())
print(p_df.info())
```

```python
X = p_df.drop(['status', 'name'], axis=1)
y = p_df['status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
from xgboost import XGBClassifier

# Remove the 'use_label_encoder' parameter, as it's not needed in the latest versions of XGBoost
xgb_model = XGBClassifier(eval_metric='logloss')
xgb_model.fit(X_train, y_train)

# Make predictions and calculate accuracy for XGBoost
y_pred_xgb = xgb_model.predict(X_test)

# Accuracy for XGBoost
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")

# Classification Report and Confusion Matrix for XGBoost
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
print("XGBoost Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```
